Continuação (if else)

```typescript
const numero = 10;

if (numero >= 0){
    console.log('O número é maior ou igual a 0');
}
```

```typescript
const numero = 5;

if (numero >= 0 && numero <= 5){
    console.log('O número está entre 0 e 5');
}
```

```typescript
const numero = 10;

if (numero >= 0 && numero <= 5){
    console.log('O número está entre 0 e 5');
}else{
    console.log('O número não está entre 0 e 5');
}
```
