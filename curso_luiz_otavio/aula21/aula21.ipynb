# Aula 21

Operadores de comparação

\> maior que


```typescript
console.log(10 > 5);
```

\>= maior que ou igual a

```typescript
const comparacao = 5 >= 5;
console.log(comparacao);
```

< menor que

```typescript
const comparacao = 5 < 5;
console.log(comparacao)
```

<= menor que ou igual a

```typescript
const comparacao = 3 <= 2;
console.log(comparacao)
```

== igual (valor, não recomendado)

```typescript
const comparacao = 2 == '2';
console.log(comparacao)
```


=== igual estrito (valor e tipo, recomendado)


```typescript
const comparacao = 2 === '2';
console.log(comparacao);
```

!= diferente (valor, não recomendado)

```typescript
const comparacao = 2 != '2';
console.log(comparacao)
```

!== diferente estrito (valor e tipo, recomendado)

```typescript
const comparacao = 2 !== '2';
console.log(comparacao);
```
