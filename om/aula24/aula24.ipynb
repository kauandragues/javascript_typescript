Entre 0 - 11 (Bom dia)

Entre 12 - 17 (Boa tarde)

Emtre 18 - 23 (Boa noite)

```typescript
const hora = 24;

if (hora >= 0 && hora <= 11){
    console.log('Bom dia, Kauã');
} else if(hora >= 12 && hora <= 17){
    console.log('Boa tarde, Kauã');
} else if(hora <= 23){
    console.log('Boa noite, Kauã');
}else{
    console.log('Erro!');
}
```
