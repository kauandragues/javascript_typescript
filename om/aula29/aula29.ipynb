Switch Case

```typescript
function formatarDiaSemana(diaSemana){
    switch (diaSemana){
    case 0:
        return 'Domingo';
    case 1:
        return 'Segunda';
    case 2:
        return 'Terça';
    case 3:
        return 'Quarta';
    case 4:
        return 'Quinta';
    case 5:
        return 'Sexta';
    case 6:
        return 'Sábado';
    default:
        return '';
}
}

const data = new Date(Date.now());
const diaSemana = formatarDiaSemana(data.getDay());

console.log(`Hoje é ${diaSemana}`);
```
