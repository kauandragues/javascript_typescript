Avaliação de curto-circuito

&& -> todas as expressões precisam ser verdadeiras para que o resultado seja verdadeiro
Se uma expressão é `false`, as outras não precisam ser avaliadas, já que uma sendo falsa o resultado é falso.

Se o valor total da expressão for falso, será retornado o VALOR da expressão falsa.

Se o valor total da expressão for verdadeiro, será retornado o valor da última expressão.

## Valores falsos em JS
`false`

0

'' ou "" ou \`\` (strings vazias)

`null`

`undefined`

`Nan`

```typescript
console.log('Luiz Otávio' && 0 && 'Maria');
```

```typescript
function falaOi(){
    return 'Oi';
}

const podeExecutar = true;
console.log(podeExecutar && falaOi()); // só será possivel chamar a função se a variavel podeExecutar for true
```

```typescript
console.log('Kaua' && 3 && 'Pedro');
```

|| -> precisa de apenas uma expressão verdadeira para retornar verdadeiro

Ele vai precisar até a primeira expressão verdadeira para poder retornar verdadeiro, se não achar é falso.

Retorna o VALOR da primeira expressão verdadeira que achar.

```typescript
const comparacao = 0 || false || null || 'Kaua' || true;
console.log(comparacao);
```

```typescript
const corUsuario = null;
const corPadrao = corUsuario || 'preto';
console.log(corPadrao);
```

```typescript
const corUsuario = 'vermelho';
const corPadrao = corUsuario || 'preto';
console.log(corPadrao);
```

```typescript
const a = 0;
const b = null;
const c = 'false';
const d = false;
const e = false;
const f = NaN;

console.log(a || b || c || d || e || f);
```
