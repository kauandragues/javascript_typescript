## Operadores Lógicos

&& -> AND -> E (todas precisam ser verdadeiras para ser veidadeira)

```typescript
const estouVivo = true;
const estouAcordado = true;
const estouComFome = true;
console.log('É verdade que estou vivo, acordado e com fome?\n', estouVivo && estouAcordado && estouComFome);
```

|| -> OR -> OU (apenas uma precisa ser verdadeira para ser verdadeira)

```typescript
const gostoDeVerde = true;
const gostoDeVermelho = false;
console.log('É verdade que gosto de verde ou vermelho?\n', gostoDeVerde || gostoDeVermelho);
```

! -> NOT -> NÃO (se verdadeiro é falso, e se falso é verdadeiro)

```typescript
const estaChovendo = false;
console.log(!estaChovendo);
```

Juntando vários

```typescript
const tenho_dinheiro = true;
const esta_chovendo = false;

console.log('Se eu tenho dinheiro e não estiver chovendo vou lá fora, posso ir?\n', tenho_dinheiro && !esta_chovendo);
```
